## SVM With Gaussian Kernel
<img src='kernel.png'>

```python
def gaussianKernel(x1,x2,sigma):
    similarity = x1-x2
    norm = np.dot(similarity.transpose(),similarity)
    return(np.exp(-((norm)/(2*np.power(sigma,2)))))
```

```python
x1 = np.array([1.0, 2.0, 1.0])
x2 = np.array([0.0, 4.0, -1.0])
sigma = 2
gaussianKernel(x1,x2,sigma)
```

```python
data3['X'].shape
```

```python
X3=pd.DataFrame(data3['X'],columns=['A','B'])
X3['Target'] = data3['y']
print X3.head()
X3val=pd.DataFrame(data3['Xval'],columns=['A','B'])
X3val['Target'] = data3['yval']
print X3val.head()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X3.A[X3.Target == 0],X3.B[X3.Target == 0],s=100, c='r',marker = 'o',label='Negative')
plt.scatter(X3.A[X3.Target == 1],X3.B[X3.Target == 1],s=100, c='b',marker = '+',label='Positive')
plt.legend()
```

```python
#Create Grid Search
C = np.array([0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30])
Gamma = np.array([0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30])

from sklearn.grid_search import ParameterGrid
param_grid = {'C': C, 'Gamma' : Gamma}
grid = ParameterGrid(param_grid)
```

```python
#Search For Best Parameter
metric = {'C':[],'Gamma':[],'Score':[],'ScoreTrain':[]}
for params in grid:
    svClassifier3 = SVC(C=params['C'],gamma=params['Gamma'],kernel='rbf')
    svClassifier3.fit(X3.iloc[:,:2],X3['Target'])
    metric['C'].append(params['C'])
    metric['Gamma'].append(params['Gamma'])
    metric['Score'].append(float(svClassifier3.score(X3val.iloc[:,:2],X3val['Target'])))
    metric['ScoreTrain'].append(float(svClassifier3.score(X3.iloc[:,:2],X3['Target'])))
```

```python
metricPD = pd.DataFrame(metric)
metricPD.head(6)
```

```python
#Get Best Model
index = np.array(metric['Score']).argmax()
C_best = metric['C'][index]
Gamma_best = metric['Gamma'][index]
svClassifier3 = SVC(C=C_best,gamma=Gamma_best,kernel='rbf')
svClassifier3.fit(X3,Y3)

#Plot Decision Boundary and Vectors
plt.figure(figsize=(12,8))
plt.scatter(svClassifier3.support_vectors_[:,0],svClassifier3.support_vectors_[:,1],\
           s=150,marker = 'o',c='y',label='Support Vector')
plt.scatter(X3.A[X3.Target == 0],X3.B[X3.Target == 0],s=100, c='r',marker = 'o',label='Negative')
plt.scatter(X3.A[X3.Target == 1],X3.B[X3.Target == 1],s=100, c='b',marker = '+',label='Positive')


plt.xlabel('X')
plt.ylabel('Y')
#plt.ylim(0.4,1)
plt.legend()
```

```python
data4= loadmat(path+'\spamTrain.mat')
data4test = loadmat(path+'\spamTest.mat')
print data4.keys()
print data4test.keys()
```

```python
spamTrain = pd.DataFrame(data4['X'])
spamTrain['Spam'] = data4['y'] 
spamTest = pd.DataFrame(data4test['Xtest'])
spamTest['Spam'] = data4test['ytest'] 

print spamTest.shape, spamTrain.shape
```

```python
spamTrain.head()
```

```python
svcSpam = SVC(C=0.1, kernel = 'linear')
svcSpam.fit(spamTrain.iloc[:,:-1], spamTrain.Spam)
print('Training accuracy = {0}%'.format(np.round(svcSpam.score(spamTrain.iloc[:,:-1], spamTrain.Spam) * 100, 2)))
print('Test accuracy = {0}%'.format(np.round(svcSpam.score(spamTest.iloc[:,:-1], spamTest.Spam) * 100, 2)))
```

## Top Predictors of Spam

```python
vocab = scipy.loadtxt(path+'/vocab.txt', dtype='str')[:,1]
vocab.shape
```

```python
top_pred_index = np.argsort(svcSpam.coef_,axis=None)[::-1]
```

```python
word_importance = {'Word':[],'Weight':[]}
for i in top_pred_index:
    word_importance['Word'].append(vocab[i])
    word_importance['Weight'].append(svcSpam.coef_.ravel()[i])

word_importancePD = pd.DataFrame(word_importance)
print "Top 20 Spam Predictors"
print word_importancePD.head(20)
print "Bottom 20 Spam Predictors"
print word_importancePD.tail(20)
```
