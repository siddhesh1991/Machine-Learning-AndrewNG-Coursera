## Cost<img src="Cost.png">

```python
def computeCost(weight,X,Y,isgrad = False) :
    y1 = Y*np.log(sigmoid(np.dot(X,weight)))
    y0 = (1-Y)*np.log(1-sigmoid(np.dot(X,weight)))
    cost = np.sum(-y1-y0)/len(X)
    if isgrad == False:
        return cost
    else:
        grad = gradient(weight, X, Y)
        return cost,grad
```

```python
#Create Matrix of Data
data.insert(0,'Intercept',1)
X=data.values[:,:3]
Y=data.Target.values
weight = np.zeros(3)
```

```python
computeCost(weight,X,Y)
```

## Gradient <img src="Gradient.png">

```python
def gradient(weight,X,Y):
    hx = sigmoid(np.dot(X,weight))
    error = hx - Y
    grad = np.dot(X.transpose(),error)/len(X)
    return grad
gradient(weight, X, Y)
```

## Use Scipy Optimization algorithms to minimize cost and get weights for Logistic Regression

```python
result1 = optimize.fmin_tnc(func=computeCost, x0=weight, fprime=gradient, args=(X, Y))
result2 = optimize.minimize(fun = computeCost,x0 = weight,args =(X,Y),method = None,jac =gradient,options={'maxiter':400})
isgrad = True
result3 = optimize.minimize(computeCost,weight, method='L-BFGS-B', args=(X, Y,isgrad), jac=True)
print 'Result1: \n',result1[0],'\n','Cost: \n',computeCost(np.array(result1[0]),X,Y),'\n'
print 'Result2: \n',result2.x,'\n','Cost: \n',computeCost(np.array(result2.x),X,Y),'\n'
print 'Result3: \n',result3.x,'\n','Cost: \n',computeCost(np.array(result3.x),X,Y),'\n'
```

```python
def predict(weight, X,threshold):
    probability = sigmoid(np.dot(X,weight))
    return [1 if x >= threshold else 0 for x in probability]
```

```python
prediction = predict(np.array(result1[0]), X,0.5)
correct = [1 if ((a == 1 and b == 1) or (a == 0 and b == 0)) else 0 for (a, b) in zip(prediction, Y)]
accuracy = (sum(map(int, correct)) % len(correct))
print 'Accuracy = {}%'.format(accuracy)
```

## Cost With Regularization <img src="CostReg.png">

```python
def computeCostReg(weight,X,Y,penalty) :
    y1 = Y*np.log(sigmoid(np.dot(X,weight)))
    y0 = (1-Y)*np.log(1-sigmoid(np.dot(X,weight)))
    cost = np.sum(-y1-y0)/len(X)
    reg = (penalty/(2*len(X)))*np.sum(np.power(weight[1:],2))
    costReg = cost + reg
    return costReg
```

## Gradient for Regularization <img src='GradientUp.png'>

```python
def gradientDescentReg(X,Y,weight,alpha,iters,penalty):
    cost = np.zeros(iters)
    for j in range(iters):
        error = (sigmoid(np.dot(X,weight)) - Y) #compute error and keep it same for each weight update for simultaneous update
        for i in range(len(weight)):
            effect = np.dot(X[:,i].transpose(),error)
            if i == 0:
                dW = (effect/len(X)) 
            else:
                dW = (effect/len(X)) + ((penalty / len(X)) * weight[i])
            weight[i] = weight[i] - (alpha*dW)
        cost[j] = computeCostReg(weight,X,Y,penalty)
    return weight,cost

#Calculate Weights with Gradient Descent - Regularized,Vecotorized
def gradientDescentRegVec(X,Y,weight,alpha,iters,penalty):
    cost = np.zeros(iters)
    for j in range(iters):
        error = (sigmoid(np.dot(X,weight)) - Y) #compute error and keep it same for each weight update for simultaneous update
        effect = np.dot(X.transpose(),error)
        dW = (effect/len(X)) + ((penalty / len(X)) * weight)
        dW[0] = (effect[0]/len(X)) 
        weight = weight - (alpha*dW)
        cost[j] = computeCostReg(weight,X,Y,penalty)
    return weight,cost
```

```python
def gradientReg(weight, X, Y, penalty):
    grad = np.zeros(len(weight))
    error = (sigmoid(np.dot(X,weight)) - Y)
    for i in range(len(weight)):
        effect = np.dot(X[:,i].transpose(),error)#np.dot(error, X[:,i]) #or 
        if (i == 0):
            grad[i] = effect / len(X)
        else:
            grad[i] = (effect / len(X)) + ((penalty / len(X)) * weight[i])
    return grad

def gradientRegVec(weight, X, Y, penalty):
    error = (sigmoid(np.dot(X,weight)) - Y)
    effect = np.dot(X.transpose(),error)#np.dot(error, X[:,i]) #or 
    grad = (effect / len(X)) + ((penalty / len(X)) * weight)
    grad[0] = effect[0] / len(X)
    return grad
```

## Supply Cost and Gradient to Scipys Optimize function to calculate optimal parameters

```python
result_reg = optimize.fmin_tnc(func=computeCostReg, x0=weight_r, fprime=gradientReg, args=(X_r, Y_r, penalty))
```

```python
print 'Weights From Scipy Optimization: \n',result_reg[0],'\n'
print 'Cost :',computeCostReg(result_reg[0], X_r, Y_r,penalty)
```

```python
result_reg2 = optimize.minimize(computeCostReg, weight_r, method='L-BFGS-B',
               jac=gradientReg, args=(X_r, Y_r, penalty),
               options={'gtol': 1e-4, 'disp': False, 'maxiter': 1000})
```

```python
print 'Weights From Scipy Optimization: \n',result_reg2.x,'\n'
print 'Cost :',computeCostReg(result_reg2.x, X_r, Y_r,penalty)
```

```python
weight_scipy = np.array(result_reg[0])
weight_scipy2 = np.array(result_reg2.x)
weight_grad = new_weight_r

predictions_scipy = predict(weight_scipy, X_r,0.5)
predictions_scipy2 = predict(weight_scipy2, X_r,0.5)
predictions_grad = predict(weight_grad, X_r,0.5)

correct_scipy = [1 if ((a == 1 and b == 1) or (a == 0 and b == 0)) else 0 for (a, b) in zip(predictions_scipy, Y_r)]
correct_scipy2 = [1 if ((a == 1 and b == 1) or (a == 0 and b == 0)) else 0 for (a, b) in zip(predictions_scipy2, Y_r)]
correct_grad = [1 if ((a == 1 and b == 1) or (a == 0 and b == 0)) else 0 for (a, b) in zip(predictions_grad, Y_r)]

accuracy_scipy = round(sum(correct_scipy)/float(len(correct_scipy)),3)*100
accuracy_scipy2 = round(sum(correct_scipy2)/float(len(correct_scipy2)),3)*100
accuracy_grad = round(sum(correct_grad)/float(len(correct_grad)),3)*100

print 'Accuracy Scipy = {}%'.format(accuracy_scipy)
print 'Accuracy Scipy 2 = {}%'.format(accuracy_scipy2)
print 'Accuracy Gradient = {}%'.format(accuracy_grad)
```
