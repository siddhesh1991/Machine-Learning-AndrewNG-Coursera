```python
def findClosestCentroids(DATA,Centroid):
    closest_centroid = []
    for i in DATA:
        distance = []
        for j in Centroid:
            dist = np.sqrt(np.dot(i-j,i-j))
            distance.append(dist)
        closest_centroid.append(np.array(distance).argmin())
    return np.array(closest_centroid)
```

```python
K = 3
initial_centroids = np.array([[3,3],[6,2],[8,5]])
findClosestCentroids(X,initial_centroids)[:3]
```

```python
def computeCetroid(X,idx,K):
    centroid =[]
    for i in range(K):
        centroid.append(np.mean(X[idx == i],axis=0))
    return np.array(centroid)
```

```python
idx = findClosestCentroids(X,initial_centroids)
computeCetroid(X,idx,K)
```

```python
def runKmeans(DATA,K,init_centroids,maxiter):
    centroid = init_centroids
    for i in range(maxiter):
        idx = findClosestCentroids(DATA,centroid)
        centroid = computeCetroid(DATA,idx,K)
    return idx,centroid
```

## Kmeans Clustering

```python
K = 3
initial_centroids = np.array([[3,3],[6,2],[8,5]])
maxiter = 10
idx,centroid  = runKmeans(X,K,initial_centroids,maxiter)

plt.figure(figsize=(12,8))
plt.scatter(X[:,0],X[:,1],s=100, c=idx,cmap=plt.cm.prism)
plt.scatter(centroid[:,0], centroid[:,1], marker='+', s=100, c='k')
plt.title("Clusters")
```

```python
def kMeansInitCentroid(DATA,K):
    index = np.arange(len(DATA))
    np.random.shuffle(index)
    centroid = DATA[index[:K],:]
    return centroid
```

```python
#Random Centroids
K = 3
initial_centroids = kMeansInitCentroid(X,K)
maxiter = 10
idx,centroid  = runKmeans(X,K,initial_centroids,maxiter)
plt.figure(figsize=(12,8))
plt.scatter(X[:,0],X[:,1],s=100, c=idx,cmap=plt.cm.prism)
plt.scatter(centroid[:,0], centroid[:,1], marker='+', s=100, c='k')
plt.title("Clusters")
```
